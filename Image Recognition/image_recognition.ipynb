```python
import face_recognition
import os

def find_similar_faces(target_image_path, images_folder):
    """
    Encontra as imagens mais semelhantes à imagem alvo em uma pasta de imagens.
    
    Args:
    - target_image_path (str): Caminho para a imagem alvo.
    - images_folder (str): Caminho para a pasta contendo as imagens a serem comparadas.
    
    Returns:
    - List[Tuple[str, float]]: Lista de tuplas contendo o caminho da imagem e o similarity score.
    """
    # Carregar a imagem alvo e codificar o rosto
    target_image = face_recognition.load_image_file(target_image_path)
    target_encoding = face_recognition.face_encodings(target_image)
    
    if not target_encoding:
        print("Nenhum rosto encontrado na imagem alvo.")
        return []
    
    target_encoding = target_encoding[0]  # Assume-se que há apenas um rosto na imagem alvo.
    
    results = []
    
    # Iterar por todas as imagens na pasta
    for image_name in os.listdir(images_folder):
        image_path = os.path.join(images_folder, image_name)
        
        # Ignorar arquivos não suportados
        if not image_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            continue
        
        # Carregar a imagem atual e codificar o rosto
        current_image = face_recognition.load_image_file(image_path)
        current_encodings = face_recognition.face_encodings(current_image)
        
        # Comparar cada rosto encontrado na imagem atual
        for encoding in current_encodings:
            # Calcular a distância entre os rostos
            face_distance = face_recognition.face_distance([target_encoding], encoding)[0]
            
            # Converter a distância para similaridade
            similarity_score = 1 - face_distance  # Maior score = mais semelhante
            
            results.append((image_name, similarity_score))
    
    # Ordenar resultados pela maior similaridade
    results = sorted(results, key=lambda x: x[1], reverse=True)
    
    return results


# --- Exemplo de Uso ---
# Caminho da imagem alvo (Input1)
target_image_path = "Imagem Alvo\\eu.jpeg"

# Caminho da pasta com imagens (Input2)
images_folder = "Varias Imagens"

# Encontrar os rostos mais semelhantes
similar_faces = find_similar_faces(target_image_path, images_folder)

# Exibir os resultados
if similar_faces:
    print("Fotos mais semelhantes:")
    for image_name, score in similar_faces:
        print(f"Imagem: {image_name}, Similarity Score: {score:.4f}")
else:
    print("Nenhuma semelhança encontrada ou nenhum rosto detectado.")
```
